```python
#Lists
fruits = ["apple", "banana", "cherry"]
fruits.append("orange")
for fruit in fruits:
    print(fruit)
```

```python
#Tuples
colors = ("red", "green", "blue")
print(colors[0])
print(colors[2])
for color in colors:
    print(color)
```

```python
#sets
numbers = {1, 2, 3, 4}
numbers.add(5)
numbers.remove(2)
for num in numbers:
    print(num)
```

```python
#3. Write the code to download, save, and transform the dataset from a given URL.
import pandas as pd

# URL of dataset (example CSV)
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv"

# Download data
df = pd.read_csv(url)

# Save dataset locally
df.to_csv("iris_dataset.csv", index=False)

print("Dataset downloaded and saved successfully!")

# ---- Transformations ----
# Display first 5 rows
print(df.head())

# Add a new column (example transformation)
df["sepal_area"] = df["sepal_length"] * df["sepal_width"]

# Save transformed data
df.to_csv("iris_transformed.csv", index=False)

print("Dataset transformed and saved!")
```

```python
'''4.Write the code to create, save, and read a file from the local drive and
perform text operations such as counting the number of lines, words, and
characters.'''
# Create and write to a file
file_name = "sample_text.txt"

with open(file_name, "w") as file:
    file.write("Hello World\n")
    file.write("This is a Python file handling example\n")
    file.write("Counting lines words and characters")

print("File created and data written!")

# Read file and perform operations
with open(file_name, "r") as file:
    text = file.read()

# Count lines
lines = text.split("\n")
line_count = len(lines)

# Count words
word_count = len(text.split())

# Count characters
char_count = len(text)

print("Number of lines:", line_count)
print("Number of words:", word_count)
print("Number of characters:", char_count)
```
