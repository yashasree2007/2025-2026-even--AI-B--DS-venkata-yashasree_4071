```python
#implement the Lasso Model
from sklearn.linear_model import Lasso
import numpy as np
X=np.array([[1,25,160],[2,30,165],[3,35,170]])
y=np.array([30000,40000,50000])
model=Lasso(alpha =0.1)
model.fit(X,y)
print(model.coef_)
```

```python
#Implement the similarity and distances for any data set
import numpy as np
from scipy.spatial import distance
from scipy.stats import pearsonr
p1=np.array([2,3])
p2=np.array([5,7])
euc = distance.euclidean(p1,p2)
mh = distance.cityblock(p1,p2)
cb = distance.chebyshev(p1,p2)
hm= distance.cosine(p1,p2)
cs = 1-cb
set1,set2 = set(p1),set(p2)
js = len(set1&set2)/len(set1|set2)
r =pearsonr(p1,p2)
print("Euclidean distance",euc)
print("Manhattan distance",mh)
print("chebyshev distance",cb)
print("Hamming distance",hm)
print("cosine similarity",cs)
print("jaccard similarity",js)
print("pearsonr's coefficent",r)
```

```python
#write a code to implement the K-means clustering(Unsupervised learning) algorithm
import numpy as np
A = np.array([[2,3],[3,4],[4,2]])
B = np.array([[8,7],[9,6],[10,8]])
P = np.array([5,5])
CA = A.mean(axis = 0)
CB = B.mean(axis = 0)
dA = np.linalg.norm(P-CA)
dB = np.linalg.norm(P-CB)
if dA<dB:
    print("point belongs to cluster A")
else:
    print("point belongs to cluster B")
```
