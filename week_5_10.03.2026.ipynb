```python
#z-score
import numpy as np
data = [10,20,30,40,50]
mean = np.mean(data)
print("Mean:",mean)
std_dev = np.std(data,ddof = 1 )
print("Standard deviation:",std_dev)
x_mean=40
z_score =(x_mean)/std_dev
print("z_score of 40:",z_score)
z_scores = [(x_mean)/std_dev for x in data]
print("z_scores:",z_scores)
```
